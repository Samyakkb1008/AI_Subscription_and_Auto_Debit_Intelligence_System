```python
import pandas as pd
import numpy as np
from datetime import timedelta

# Load cleaned transactions dataset
df = pd.read_csv("..\\data\\subscription_dataset_cleaned.csv")
df["Date"] = pd.to_datetime(df["Date"])

df.head()
```

```python
df["SubscriptionFlag"].value_counts()
```

**Compute Date Intervals**

```python
""" Calculate day differences between consecutive dates.
    Used to detect weekly / monthly patterns.
"""
def compute_intervals(dates):
    dates = sorted(dates)  # Ensure chronological order
    intervals = [
        (dates[i] - dates[i-1]).days
        for i in range(1, len(dates))
    ]
    return intervals
```

**Determine Recurrence Type (Weekly / Monthly)**

```python
""" Classifies recurrence based on average interval:
    - Weekly: 5 to 9 days
    - Monthly: 28 to 32 days
"""
def classify_recurrence(intervals):
    
    if len(intervals) < 2:
        return None  # Not enough data

    avg_gap = np.mean(intervals)

    if 5 <= avg_gap <= 9:
        return "WEEKLY"
    if 28 <= avg_gap <= 32:
        return "MONTHLY"

    return None
```

**Detect Recurring Patterns + Next Date**

```python
from dateutil.relativedelta import relativedelta


patterns = []

for customer in df["CustomerID"].unique():
    df_cust = df[df["CustomerID"] == customer]

    for merchant in df_cust["Merchant"].unique():
        df_cm = df_cust[df_cust["Merchant"] == merchant]

        # Need at least 3 occurrences to consider recurring
        if len(df_cm) < 3:
            continue

        # Extract dates
        dates = sorted(df_cm["Date"].tolist())
        intervals = compute_intervals(dates)

        # Determine recurrence type
        recurrence = classify_recurrence(intervals)
        if recurrence is None:
            continue  # Not recurring

        # Recurring day (e.g., 5th of every month)
        recurring_day = df_cm["Date"].dt.day.mode()[0]

        # Predict next expected date
        last_date = dates[-1]

        if recurrence == "WEEKLY":
            next_expected = last_date + timedelta(days=7)
        elif recurrence == "MONTHLY":
            # Correct month-handling including Feb and month-end cases
            next_expected = last_date + relativedelta(months=1)

        # elif recurrence == "MONTHLY":
        #     next_expected = last_date + timedelta(days=30)  # or use avg interval

        patterns.append({
            "CustomerID": customer,
            "Merchant": merchant,
            "Recurrence": recurrence,
            "RecurringDay": int(recurring_day),
            "LastTransactionDate": last_date.date(),
            "NextExpectedDate": next_expected.date(),
            "TotalOccurrences": len(df_cm)
        })
```

**Convert to DataFrame**

```python
patterns_df = pd.DataFrame(patterns)
patterns_df
```

**Save for ML / Dashboard**

```python
patterns_df.to_csv("..\\data\\recurring_patterns.csv", index=False)
patterns_df
```
