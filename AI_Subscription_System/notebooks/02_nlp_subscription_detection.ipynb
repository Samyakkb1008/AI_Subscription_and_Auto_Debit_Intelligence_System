## **Building NLP Model**
### **NLP Subscription Detection**
- **Objective:**
  - To classify transactions as subscription or non-subscription using NLP and machine learning techniques.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
import pickle
```

We used spaCy for keyword-based merchant extraction and combined it with a machine learning classifier to identify subscription transactions.

### **1: Extract Merchant using spaCy**

```python
import spacy

nlp = spacy.load("en_core_web_sm")

# Predefined merchant list (important)
merchant_list = ["netflix", "spotify", "amazon", "electricity", "hotstar"]

def extract_merchant(text):
    text = str(text).lower()
    
    for merchant in merchant_list:
        if merchant in text:
            return merchant.capitalize()
    
    return "Unknown"

df["Merchant_Extracted"] = df["Description"].apply(extract_merchant)
```

### **2: Text Classification (Subscription or Not)**

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

# Label creation
df["label"] = df["Merchant_Extracted"].apply(lambda x: 1 if x != "Unknown" else 0)

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df["Description"])
y = df["label"]

model = LogisticRegression()
model.fit(X, y)
```

### **3: Save Model**

```python
# Create Labels
keywords = ["netflix", "spotify", "amazon prime", "hotstar", "bill", "recharge"]

def label_data(text):
    text = str(text).lower()
    for word in keywords:
        if word in text:
            return 1   # Subscription
    return 0          # Non-subscription

df["label"] = df["Description"].apply(label_data)
```
